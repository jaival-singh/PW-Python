Q1. Explain why we have to use the Exception class while creating a Custom Exception.

In Python, custom exceptions are defined by creating a new class that inherits from the Exception class or one of its subclasses. The reason for inheriting from the Exception class is that it provides a standardized interface and behavior for all exceptions in Python.In addition, inheriting from the Exception class allows your custom exception to be caught by any try/except block that catches the base Exception class or one of its subclasses. This makes it easier to handle and manage exceptions in your code.

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

The ArithmeticError class is a base class for all errors that occur during arithmetic operations. It has several subclasses that represent specific types of arithmetic errors.

ZeroDivisionError: This error is raised when a number is divided by zero.

```python
try:
    result = 1 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")  
```

OverflowError: This error is raised when the result of a floating-point operation is too large to be represented as a finite value.

```python
import sys

try:
    result = sys.maxsize + 1
except OverflowError as e:
    print(f"Error: {e}")  
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is a base class for all errors that occur when a key or index is not found in a sequence or mapping. It has two common subclasses: KeyError and IndexError.

KeyError: This error is raised when you try to access a key in a dictionary that does not exist.

```python
my_dict = {"a": 1, "b": 2}

try:
    value = my_dict["c"]
except KeyError as e:
    print(f"Error: {e}")  
```

IndexError: This error is raised when you try to access an index in a sequence that does not exist.

```python
my_list = [1, 2, 3]

try:
    value = my_list[3]
except IndexError as e:
    print(f"Error: {e}")  
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is an exception that is raised when an imported module or a submodule cannot be found or imported successfully. This error can occur for several reasons, such as:

The module or submodule does not exist.
The module or submodule is not installed.
There is a problem with the module's code or dependencies.

```python
try:
    import non_existent_module
except ImportError as e:
    print(f"Error: {e}") 
```

ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6. It is raised when a module or package cannot be found or imported. The main difference between ImportError and ModuleNotFoundError is that the latter is more specific and provides a more detailed error message that includes the name of the module that could not be found.

```python
try:
    import non_existent_module
except ModuleNotFoundError as e:
    print(f"Error: {e}")  
```
